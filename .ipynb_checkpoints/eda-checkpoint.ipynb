 Cuisine and Restaurant Type Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the most common cuisines (considering only the first listed cuisine if multiple are listed)
df['Primary Cuisine'] = df['Cuisines'].str.split(',').str[0]

# Get top 10 most common cuisines
top_cuisines = df['Primary Cuisine'].value_counts().head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette="viridis")
plt.title("Top 10 Most Common Cuisines")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine")
plt.tight_layout()
plt.show()
```

```python
# Group by primary cuisine and table booking availability
booking_cuisine = df.groupby(['Primary Cuisine', 'Has Table booking']).size().unstack().fillna(0)

# Plot
booking_cuisine = booking_cuisine.loc[booking_cuisine.sum(axis=1).sort_values(ascending=False).head(10).index]

booking_cuisine.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='coolwarm')
plt.title("Table Booking Trends by Top 10 Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group by primary cuisine
cuisine_stats = df.groupby('Primary Cuisine')[['Aggregate rating', 'Average Cost for two']].mean().dropna()

# Sort by rating
cuisine_stats_sorted = cuisine_stats.sort_values(by='Aggregate rating', ascending=False).head(10)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=cuisine_stats_sorted.reset_index(), x='Primary Cuisine', y='Aggregate rating', palette='Blues_d')
plt.title("Top 10 Cuisines by Average Rating")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot average cost
cuisine_stats_sorted_cost = cuisine_stats.sort_values(by='Average Cost for two', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(data=cuisine_stats_sorted_cost.reset_index(), x='Primary Cuisine', y='Average Cost for two', palette='Purples_d')
plt.title("Top 10 Cuisines by Average Cost for Two")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Rating and Pricing Behavior

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Aggregate rating'], bins=15, kde=True, color='skyblue')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Aggregate rating', y='Votes', alpha=0.5, hue='Rating color', palette='Set1')
plt.title('Votes vs. Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Votes')
plt.legend(title='Rating Color')
plt.tight_layout()
plt.show()

# Correlation coefficient
correlation = df[['Aggregate rating', 'Votes']].corr().iloc[0,1]
print(f"Correlation between Votes and Aggregate Rating: {correlation:.2f}")
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Price range', y='Aggregate rating', palette='coolwarm')
plt.title('Aggregate Rating by Price Range')
plt.xlabel('Price Range (1 = Low, 4 = High)')
plt.ylabel('Aggregate Rating')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Aggregate rating', y='Average Cost for two', scatter_kws={'alpha':0.4}, line_kws={"color":"red"})
plt.title('Do Higher Ratings Mean Higher Costs?')
plt.xlabel('Aggregate Rating')
plt.ylabel('Average Cost for Two')
plt.tight_layout()
plt.show()

# Correlation coefficient
corr_cost_rating = df[['Aggregate rating', 'Average Cost for two']].corr().iloc[0,1]
print(f"Correlation between Rating and Cost for Two: {corr_cost_rating:.2f}")
```

```python
from collections import Counter

cuisine_list = df['Cuisines'].dropna().str.split(', ')
flat_cuisines = [cuisine for sublist in cuisine_list for cuisine in sublist]
cuisine_counts = Counter(flat_cuisines).most_common(10)

# Convert to DataFrame for plotting
cuisine_df = pd.DataFrame(cuisine_counts, columns=['Cuisine', 'Count'])
plt.figure(figsize=(10,5))
sns.barplot(data=cuisine_df, y='Cuisine', x='Count', palette='rocket')
plt.title("Top 10 Most Common Cuisines")
plt.show()
```

```python
import altair as alt

cuisine_list = df['Cuisines'].dropna().str.split(', ')
flat_cuisines = [c for sublist in cuisine_list for c in sublist]
cuisine_counts = pd.Series(flat_cuisines).value_counts().reset_index().head(10)
cuisine_counts.columns = ['Cuisine', 'Count']

chart = alt.Chart(cuisine_counts).mark_bar().encode(
    x='Count:Q',
    y=alt.Y('Cuisine:N', sort='-x'),
    tooltip=['Cuisine', 'Count']
).properties(
    width=600,
    height=400,
    title='Top 10 Cuisines'
)

chart.display()
```
